# NP array attributes

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

a3
```

```python
# ndim
a3.ndim
```

```python
# shape
print(a3.shape)
a3
```

```python
# size
print(a2.size)
a2
```

```python
# itemsize
a3.itemsize
```

```python
# dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

# Changing datatype

```python
a3 = np.arange(8).reshape(2,2,2)
a3
# astype
a3.astype(np.int32)
```

# Array Operation

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)

a2
```

```python
# scalar operations

# arithmetic
a1 ** 2
```

```python
# relational
a2 == 15
```

```python
# vector operations
# arithmetic
a1 ** a2
```

# Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
# axis is used for row wise or column wise calculation
np.prod(a1,axis=0)
```

```python
# mean/median/std/var
np.var(a1,axis=1)
```

```python
# trigonomoetric functions
np.sin(a1)
```

```python
# dot product
# column of first must be equal to row of second
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents
np.exp(a1)
```

```python
# round/floor/ceil

np.ceil(np.random.random((2,3))*100)
```

# Indexing and Slicing

```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

a3
```

```python
a2[0:2,1::2]
```

```python
a2[::2,1::2]
```

```python
a3[::2,0,::2]
```

# Iteration

```python
for i in a1:
  print(i)
```

```python
for i in a2:
  print(i)
```

```python
for i in a3:
  print(i)
```

```python
# converts multi-D array into 1D first

for i in np.nditer(a3):
  print(i)
```

# Reshaping

```python
# Transpose
# Reshape has been coverd above
np.transpose(a2)
a2.T
```

```python
# ravel
a3.ravel()
```

# Stacking

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
# Horizontal stacking
np.hstack((a4,a5))
```

```python
# Vertical stacking
np.vstack((a4,a5))
```

# Spliting

```python
np.hsplit(a4,2)
```

```python
np.vsplit(a4,1)
```
